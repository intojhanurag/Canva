```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame (ensure this path is correct)
df_copy = pd.read_csv(r"C:\Users\aojha\OneDrive\Attachments\python\6-file Handline\Algerian_forest_fires_dataset.csv")  # Change to your file path

# Check if the DataFrame is loaded correctly
print(df_copy.head())  # This should display the first few rows of your DataFrame

# Check available styles
print(plt.style.available)  # List available styles

# Setting the style and plotting
try:
    plt.style.use('seaborn')  # Ensure 'seaborn' is available
except ValueError as e:
    print(f"Error setting style: {e}")
    plt.style.use('ggplot')  # Fallback to another style if seaborn is not available

df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
import matplotlib.pyplot as plt

# Sample data
classlabels = ["Fire", "Not Fire"]
 # Ensure the percentages sum to 100

# Plotting the pie chart
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')  # Note the double '%%' for percentage
plt.title("Pie Chart of Classes")
plt.show()
```
