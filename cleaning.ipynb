```python
df.columns=df.columns.str.strip()
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.dropna(subset=['month', 'day', 'year', 'Temperature', 'RH', 'Ws'], inplace=True)
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(float).astype(int)
```

```python
df_cleaned = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].apply(pd.to_numeric, errors='coerce')
df_cleaned = df_cleaned.dropna()  # Drop rows that have NaNs after conversion
df_cleaned = df_cleaned.astype(int)  # Now safely convert to int
```

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
for i in objects:
    if i != 'Classes':
        # Find non-numeric values in the column
        non_numeric = pd.to_numeric(df[i], errors='coerce').isna()
        if non_numeric.any():
            print(f"Non-numeric values found in column {i}:")
            print(df[i][non_numeric])
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = pd.to_numeric(df[i], errors='coerce')
```

```python
df.to_csv('Algerian_forest_fires_dataset.csv',index=False)
```
