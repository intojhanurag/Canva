```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)

plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title("Basic line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
```
